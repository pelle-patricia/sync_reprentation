## Curvas de sintonía en mamíferos

### Principales hallazgos biológicos

#### Relación posición-frecuencia:
 
Los estudios sobre la manera en que los sonidos se codifican en el sistema auditivo humano comienzan con el estudio del transductor de sonidos de los mamíferos: la coclea. Integrado dentro del hueso del cráneo, es un sensor diminuto y delicado de forma tubular pero enrollado sobre sí mismo de manera similar al de un caracol, de donde proviene el nombre coclea. Una membrana se despliega en su interior también longitudinalmente que es capaz de captar las más ínfimas vibraciones de las ondas de presión sonoras exteriores y reproducirlas sobre su superficie, usando las increibles adaptaciones provistas por el oido externo y medio. The manner in which this membrane vibrates in response to sound is the key to understanding cochlear function.  The most significant contribution to the understanding of this phenomenum was made by Georg von [Békésy], (<a href="#zwislocki_jasa80">zwislocki_jasa80</a>) showing that this membrane vibrates maximally at *different positions* sobre su longitud as a function of the *stimulus frequency*. Las frecuencias de estimulación de cada partición de la cóclea están distribuidas ordenadamente a lo largo de la misma, con las más graves en el extremo más interno, mientras que los sonidos más agudos son observados hacia el extremo más externo, debido a las variaciones de masa, dimensión y flexibilidad de la membrana a lo largo de su longitud. Esta descripción motiva el termino de *distribución tonotópica* del sonido o *frequency-place relation*, ya que hay un mapeo pseudo lineal entre la posición de la partición de la membrana de máxima amplitud del movimiento y la frecuencia del estímulo. (véase <a href="#Greenwood">Greenwood</a> para empezar)

La membrana tiene adosado sobre su superficie un complejo conjunto de células llamado órgano de Corti <a href="#ruggero_inbook92">ruggero_inbook93</a>, que es capaz de generar descargas neuronales en las células nerviosas que hacen sinapsis en la zona de cada partición. Mediante unas células especializadas llamadas células ciliadas capaces de realizar descargas nerviosas cada vez que hay un movimiento en su entorno, todo el conjunto neuronal-mecánico logra convertir un desplazamiento mecánico en la membrana en un impulso eléctrico que se transmite a través de las fibras del nervio auditivo al siguiente núcleo del sistema auditivo. Hay también un orden dentro de las fibras del nervio auditivo, ya que diferentes fibras están innervadas por diferentes particiones de la membrana. De este modo, hay una correspondencia de tres partes entre las frecuencias de un sonido, la posición de la máxima vibración sobre la membrana y la fibra con máxima respuesta dentro del nervio auditivo. <a href="#kiang_mitpress65">kiang_mitpress65</a>


#### Descripción detallada de la respuesta en frecuencia de las particiones de la membrana, curva de sintonía:

A medida que se fueron haciendo más y más observaciones, aumentó el detalle sobre el movimiento de las particiones cocleares. El movimiento de cada partición is typically described as a function of stimulus frequency. This yields a mechanical filter characteristic that is often called its transfer function (TF), (<a href="#Rhode">Rhode</a>) The TF consists of an amplitude and phase characteristic and has the general features of a bandpass filter. Esta transferencia suele denominarse *curva de sintonía*, nombre que pone de manifiesto los límites reales del esquema idealizado de relación frequency-place, y es derivado del ámbito de las comunicaciones, i.e., los receptores de radio frecuencia. La denominación frecuency-place relation puede inducirnos erróneamente a imaginar que una sola partición de la membrana puede responder cuando hay un estímulo de una cierta frecuencia, y que esa frecuencia es la única a la cual esa partición responde. La curva de sintonía nos muestra los límites reales de esa idealización, poniendo en juego también la variable del nivel sonoro del estímulo.  

 El relevamiento de la transferencia para una dada partición de la cóclea en su forma más simple se realiza estimulando con sonidos senoidales de diferentes frecuencias y a diferentes niveles sonoros, y al mismo tiempo observando si se produce un desplazamiento o no en esa partición (<a href="#"></a>). Existe una frecuencia de estimulo para la cual el nivel sonoro necesario para producir un movimiento dado es el mínimo posible. Esta frecuencia es llamada frecuencia característica (CF) de la partición. Si la frecuencia del estímulo cambia dentro de un rango estrecho de frecuencias, también es posible obtener desplazamientos en la membrana en la misma partición, pero siempre a un nivel sonoro mayor al utilizado para la frecuencia característica. Este rango de frecuencias es más amplio cuanto mayor es el nivel sonoro. La transferencia entonces es el umbral de nivel sonoro requerido a cada frecuencia para estimular la partición. Las vibraciones en una partición son más perceptibles a frecuencias menores a la característica que a frecuencias mayores. Es decir, que dado un nivel constante de sonido, las frecuencias mayores a CF rápidamente dejan de estimular la partición, mientras que a frecuencias menores es posibles apartarse más y seguir observando la estimulación. De este modo la transferencias son descriptas por una serie de parámetros, siendo los más importantes su frecuencia de máxima respuesta, o frecuencia característica (CF) y las pendientes de caida del umbral alrededor de esa frecuencia, que es mucho más abrupta hacia frecuencias mayores a CF que hacia las bajas frecuencias. Alternativamente se suelen usar dos parámetros relacionados a las pendientes, la agudeza y la asimetría del filtro. Las mediciones de transferencia se hicieron también para las fibras del nervio auditivo, obteniendo también similares comportamientos (<a href="#"></a>), como veremos en más detalle luego.
 
Es posible establecer un consenso sobre las curvas de sintonía entre la multiplicidad de mediciones realizadas por diversos autores en varios aspectos, ya que la variedad de especies medidas y los diferentes métodos utilizados de los experimentos a lo largo de los años dieron lugar a diferencias. Por ejemplo, un hecho en el cual hay coincidencia es que existe una correspondencia entre frecuencia característica CF del filtro mecánico y del neuronal innervado en esa partición. Otro es que en la gran mayoría de los casos la pendiente hacia altas frecuencias en ambos filtros es muy abrupta, mucho más que la pendiente hacia bajas frecuencias, es decir, las transferencias son no simétricas. Sin embargo una controversia más dificil de dirimir es que si bien tanto la forma de la transferencia de la membrana como la de la neuronal son similares, el filtro neuronal parecía más agudo que el filtro mecánico de la membrana de la cóclea, y esa diferencias era mucho más notable en la pendiente hacia bajas frecuencias. Este punto fue conflictivo no solo por las divergencias de observaciones, sino porque su origen dio lugar a numerosas interpretacione, conduciendo al cuestionamiento de si el sistema auditivo realmente se comportaba como un filtro o no (<a href="Rhode">Rhode</a>), como veremos en lo que sigue. 

#### La respuesta de las fibras del nervio audtivo: sincronía

El caso del chequeo del filtro neuronal (<a href="#Evans">Evans</a>) es similar al mecánico pero con una ligera variante. Las fibras del nervio auditivo tienen un patrón de descargas aun en reposo. La manera de medir la transferencia consiste también en variar la frecuencia del estímulo para un dado nivel de presión sonora, y detectar en qué nivel sonoro las fibras dejan su patrón de descargas de reposo para emitir el patrón de descargas de respuesta al estímulo. La diferencia entre ambos patrones de descargas es por una parte que ante estímulos hay más descargas por unidad de tiempo que en reposo y por el otro que las descargas dejan de producirse en instantes aleatorios de tiempo para pasar a ordenarse de modo sincrónico en fase con la fase del estímulo (<a href="#"></a>).  En situación de estimulación con un sonido sinusoidal puro, estas descargas ocurren mayormente en un dado tiempo dentro del período de la senoide de entrada. Este patrón de descargas sinconizadas observadas en el nervio auditivo es la principal motivación para utilizar PLLs como elementos que podrían emular la representación del sonido del sistema auditivo. 

#### Nonlinearities and the second filter hypothesis

Las diferencias entre las mediciones de la transferencia mecánica de las particiones cocleares y la transferencia neuronal de las fibras innervadas por dicha partición, hizo que se postularan una serie de hipótesis que explicaban este comportamiento. Concretamente, si la fibra del nervio auditivo está innervada  directamente a partir de una partición coclear, y ambas transferencias difieren, la pregunta es de qué modo se puede explicar en términos descriptivos ese desacuerdo. La descripción más simple posible fue esbozada en la sección anterior: si la mecánica de la partición coclear se comporta como un filtro lineal pasabanda (más propiamente, lineal e invariante ante desplazamientos, LTI), se requeriría un segundo filtro (en su forma más simple, también LTI) para explicar el aumento de agudeza de la transferencia neuronal. [ruggero_inbook92][evans_science75]

Sin embargo, dos filtros lineales colocados en cascada forman en conjunto un sistema también LTI, el cual no logra explicar satisfactoriamente la respuesta del sistema conjunto ante otros estímulos que no sean sinusoidales puros. El más claro de los experimentos es el de la supresión de dos tonos(<a href="#"></a>). Este fenómeno tiene su correlato en el siguiente experimento psicoacústico: si un individuo sano auditivamente escucha un estímulo correspondiente a un tono de un dado nivel de presión sonora, este tono puede dejar de escucharse por la presencia de otro tono de una determinada frecuencia y nivel sonoro (mayor). Este efecto de enmascaramiento de un tono por otro, se produce también por la presencia de ruido en lugar de estímulo senoidal, y tiene un correlato neuronal medible a nivel de las fibras del nervio auditivo, y también a nivel mecánico en la cóclea. Este fenómeno nos muestra un claro comportamiento no lineal, lo cual presenta un desafío para el campo de las investigaciones biológicas. Un aspecto destacable sobre el conocimiento de estos fenómenos es que el mismo ha sido aprovechado en el ámbito de las comunicaciones, en particular en la codificación, para generar la compresión MPEG de audio(<a href="#"></a>). En esta codificación se utiliza el efecto de enmascaramiento de dos tonos para seleccionar una distribución no uniforme en frecuencia del ruido agregado de cuantización. Una cuantización más ruidosa será colocada dinámicamente en las zonas donde sea más "inofensivo", es decir las zonas de influencia donde un tono de alto nivel enmascara a sus vecinos. Este tipo de codificación forma parte de los llamados códigos perceptuales(<a href="#"></a>), por utilizar estos fenómenos perceptuales en su concepción. 

## Propuesta de comportamiento utilizando PLL

En lo que sigue describiremos el sistema propuesto que podría representar los fenómenos descriptos en las secciones anteriores de una forma cualitativa: por una parte la diferencia en la transferencia mecánica y la transferencia neuronal, y por otra parte los efectos que se producen ante la presentación de combinaciones de tonos de estímulo.

En la figura vemos el diagrama de bloques de nuestro sistema, representando el efecto producido ante un estímulo en una dada partición coclear. El mismo consiste de un filtro LTI pasabanda, y un PLL colocado en cascada a su salida. En este esquema consideramos que la entrada representará un estímulo sonoro (convertido por un micrófono/amplificador en una onda de tensión, y luego muestreado a tiempo discreto), y la salida del PLL representará las descargas en las fibras del nervio auditivo de dicha partición. Esta representación no necesita ser literal, es decir, no hay reales spikes como en las fibras, pero podemos considerar que los pasos por múltiplos $2 \pi$ de la señal sinusoidal del `VCO` interno representarán los puntos donde habría descargas de la neurona. Como el aspecto más relevante representado es el sincronismo entre los estímulos y los disparos en fase de las fibras del nervio auditivo, llamaremos a este sistema *representación sincrónica*.

<a id="sinc_rep_bloques"></a>


Los detalles de implementación del PLL se explicaron en la sección anterior. La respuesta en frecuencia del filtro LTI sigue los lineamientos de (<a href="#"></a>), y tiene la forma que se muestra en el gráfico anterior, debajo del bloque del filtro. Este tipo de filtro fue propuesto siguiendo descripciones biológicas sobre la forma de las curvas de sintonía, con una frecuencia de respuesta máxima (CF), y una asimetría pronunciada, más abrupta hacia altas frecuencias. La implementación del filtro en tiempo discreto permite generar un programa que sea capaz de realizar el equivalente en tiempo discreto de las operaciones de estos dos dispositivos, inicialmente pensados en forma analógica. El pasaje de analógico a tiempo discreto en el caso del PLL fue descripto en la sección anterior. En el caso del filtro es realizado por el método de muestreo de la respuesta en frecuencia (<a href="#"></a>), obteniéndose de este modo un filtro de fase lineal ya que es un filtro de respuesta impulsiva finita (FIR, <a href="#"></a>). Este procedimiento nos permite obtener una descripción temporal de la respuesta a una señal de entrada de tipo impulsiva, con lo cual la operatoria del filtro se realizará mediante la convolución de tiempo discreto entre dicha respuesta al impulso y la señal de estímulo muestreada en tiempo. El filtro es de orden 2048, para permitir un adecuado muestreo en frecuencia, sobre todo en su pendiente abrupta. El filtro de fase lineal es intrínsecamente no causal, de modo que la respuesta del filtro siempre está retardada con respecto a la entrada en la mitad de su orden, ya que los filtros de fase lineal son además simétricos. La convolución puede ser implementada sobre segmentos consecutivos de la señal, y con adecuados recaudos por medio de multiplicaciones en el dominio del espectro discreto (DFT), de modo que las operaciones sean eficientes en tiempo de cómputo y en la cantidad de memoria utilizada, y con un corto retardo en tiempo entre el tiempo en que se produce la entrada con respecto al tiempo en que se calcula la salida final. Las ecuaciones temporales del PLL nos muestran que  este dispositivo en su versión de tiempo discreto no consume casi cómputo ni memoria, ni tiene retardos temporales, salvo de 1 muestra.

Sin embargo, un tono que se enciende y se apaga tiene una componente transitoria importante en su salida, sumada en los puntos en los que se produce el escalón de energía a la entrada. Una vez que el tono desaparece, el PLL finalmente logrará la adquisión, pero el tiempo requerido para ello puede variar.

Es importante recalcar, que el efecto de supresion es debido también a la sincronización del PLL, del mismo modo que lo era la agudeza de la transferencia resultante. Este fenómeno de sincronismo es muy robusto frente a cambios moderados, y así se mostró en las curvas de sintonía. Pero frente a cambios abruptos de la entrada, las condiciones del PLL pueden modificarse de modo que el error obtenido entre su oscilador interno y los cambios externos sea irreversible. Es decir, en estos fenómenos es necesario describir cuidadosamente no solo los aspectos estacionarios como amplitudes o frecuencias, sino el tipo de variación temporal para dichos parámetros. Cambios más abruptos en el onset de las señales son más propicios para que este efecto aparezca. Este tipo de problemas en la medición de la supresión también son descriptos en la literatura biológica experimental. 

### Mediciones ante estímulos de dos tonos no supresores. 

El experimento biológico de la supresión de dos tonos es un fenómeno donde la nolinealidad de los resultados obtenidos es evidente. Es necesario analizar también otras mediciones donde los resultados tienen un grado de linealidad y ver si este efecto es reproducible o no en la representación sincrónica propuesta. En la configuración de más a la izquierda en la figura, la sincronización se dará prioritariamente sobre el tono de frecuencia más alta, y a medida que se incrementa la frecuencia de ambos tonos la sincronización baja, ocurriendo lo contrario para el tono de más baja frecuencia. Este comportamiento es esperable dado que sabemos que las frecuencias cercanas a la característica son más susceptibles de presentar sincronismo con el estímulo. El experimento se puede utilizar para corroborar la frecuencia característica, que es el lugar donde la mitad de los disparos se sincronizan con cada uno de los tonos, y en ese caso se denomina Frequencia caracterísitca de dos tonos (2TCF). Esta frecuencia es siempre menor que la frecuencia característica medida sobre la curva de sintonía, como se muestra en este caso en la figura.

El experimento realizado para corroborar este funcionamiento en la representación sincrónica, es el análogo al experimento descripto en el párrafo anterior. Para simular la medición del vector strength en este caso vamos a generar efectivamente los disparos neuronales en los puntos de cruce por múltiplos de $2 \pi$. Es decir, generamos una señal que contiene ceros en toda su longitud, excepto en el tiempo posterior al paso de un ciclo al siguiente de la señal de salida del `VCO`, donde le asignaremos un 1. De la definición del vector strength (<a href=""></a>), podemos concluir que es posible calcularlo como la Serie de Fourier de tiempo discreto de esta señal de disparos (<a href=""></a>). Es decir, el vector strengh será proporcional al valor de amplitud de la transformada de Fourier de las salidas de disparos de las neuronas, evaluada a la frecuencia del estímulo. Esto en nuestro caso será implementado mediante una DFT, para la señal correctamente completada con ceros, de modo que se obtenga un muestreo en frecuencia discreta de la transformada de frecuencia continua, adecuado para observar los puntos en cuestión. Esta amplitud, adecuadamente normalizada, será equivalente al vector strength. La siguiente figura muestra el barrido realizado para un nivel de amplitud de 0db de ambos tonos, con frecuencias media entre ambos tonos variando desde 1700 a 2050Hz, separadas en 200Hz. Los parámetros del sistema son los mismos que en los experimentos anteriores.

Se puede ver en nuestro caso que la transición entre el sincronismo con uno u otro tono es más abrupto que en el caso anterior, pero el comportamiento guarda una semejanza aceptable con el caso biológico. En el caso de la representación sincrónica, es posible encontrar una explicación directa para varios aspectos de este experimento. Primeramente, el aumento de sincronismo del tono de más baja frecuencia, así como la disminución del otro tono, se puede explicar en la representación de manera simple, ya que frente a la aparición simultánea de dos sinusoides sumadas en su entrada, es más probable que el PLL se sincronice con la senoide de más alta energía de las dos. De acuerdo a la posición relativa de los dos tonos con respecto a la respuesta en frecuencia del filtro, uno estará más atenuado que el otro, derivando en una mayor o menor chance de sincronización. Derivado de este hecho se explica también que la frecuencia característica de dos tonos sea siempre menor que la frecuencia característica medida con un solo tono como en las curvas de sintonía, ya que la igualdad de amplitudes a la salida del filtro se da a una frecuencia media menor a la característica por la asimetría del filtro.

La frecuencia es el reflejo de $F_1$ con respecto a $F_2$, es decir, corresponde a la frecuencia $(F_2 + (F_2-F_1)) = 2 F_2 - F_1$. La aparición de esa frecuencia espejada puede entenderse de un modo bastante directo en nuestro caso, si observamos la señal temporal de frecuencia en el gráfico superior. Lo que vemos es que esta señal no es constante, sino que fluctúa en torno a la frecuencia indicada. Esto quiere decir que el `VCO` no está generando ciclos de la misma duración, sino que estos varian alrededor del período correspondiente a la energía predominante, cíclicamente. Este efecto es mucho más notable en el caso de la derecha, donde ambas senoides de estímulo tienen una amplitud comparable después de haber atravesado el filtro. Es decir, el `VCO` está generando una senoide de frecuencia modulada, no una senoide estacionaria. En ese caso, el espectro de las deltas generadas no contiene un sólo armónico, como correspondería al espectro de deltas periódico (con picos de igual amplitud en frecuencias múltiplo de la fundamental solamente), sino que presenta otras componentes de frecuencia secundarias, pero simétricas a la central, a una distancia igual a la diferencia de ambas. Este espectro corresponde no a la suma de los dos tonos, sino a la modulación en frecuencia de la senoide de mayor amplitud por otra cuya frecuencia es igual a la diferencia entre las dos:

$ = J_0(B) \cos(2 \pi F_2 t) + J_1(B) \big(\cos( 2 \pi F_1 t) + \cos( 2 \pi (2 F_2 - F_1) t)\big) $

donde se puede ver la expresión de los dos primeros armónicos que concuerda con los principales picos mostrados, y donde $J_0(B)$ y $J_1(B)$ corresponden a la amplitud de los dos primeros polinomios de Bessel evaluados en $B$. En la figura se pueden apreciar otros armónicos también, y debe recordarse que por ser un espectro de los disparos generados en comienzos de ciclos de la señal del `VCO`, el espectro se repite en múltiplos de $2 \pi F_2$. Estos armónicos serán más evidentes cuanto mayor sea el valor de $B$. De las dos figuras anteriores, vemos que este valor de $B$ es mayor cuanto mayor es la cercanía a la frecuencia de 2TCF, es decir cuanto más parecidas son las amplitudes de ambas senoides. Explicado en términos del funcionamiento del PLL, la adquisición a la señal de mayor amplitud (frecuencia $F_2$) se mantiene, pero no se puede considerar que el PLL esté cumpliendo su modelo lineal. El error en el comparador de fase es grande, mayor que lo requerido para aproximar el multiplicador a un detector lineal. Sin embargo la perturbación no es la necesaria como para determinar una pérdida de adquisión. Para concluir, es importante destacar que visto este experimento desde esta óptica de la representación de sincronía, lo que está sucediendo es complementario a lo que se concluyó de la simulación de la supresión de dos tonos. El PLL intenta mantener la adquisión a un tono predominante, pero el otro tono que produce una interferencia en este caso no tiene el cambio dinámico necesario para producir pérdida de adquisición. Sin embargo, el PLL se ve perturbado con respecto a lo que se observaría si solo un estímulo senoidal estuviera presente, y esa perturbación se expresa como una modulación en frecuencia, que también exhibe componentes no lineales. 

### Conclusiones sobre la aproximación biológica de la representación sincrónica.

Los fenómenos biológicos presentados en las secciones anteriores nos permiten concluir que la representación sincrónica es acertada en cuanto rescata un aspecto fundamental que es el sincronismo de los disparos en las fibras del nervio auditivo. El PLL fue elegido para mimetizar ese sincronismo por sus conocidas propiedades en el campo de los dispositivos electrónicos. Además de rescatar este aspecto observado en las fibras nerviosas, vimos que se puede explicar por una parte la agudización de la curva de sintonía mediante efectos relacionados al sincronismo. Por otra parte también el efecto de la supresión de dos tonos podría ser generado por un efecto de falta de adquisición dentro de un entorno sincrónico. Y por último en la medición de frecuencia característica de dos tonos, la presencia de componentes no lineales es fácilmente interpretable observando el comportamiento sincrónico, por la perturbación del sincronismo. Es necesario enfatizar además, que bajo la hipótesis de la descripción de la representación sincrónica, varios aspectos observados en los experimentos biológicos de discutida interpretación, son explicables de un modo directo y cualitativo, como es la aparición de varios tipos de no linealidades descriptas en la literatura biológica. Esta semejanza sin embargo, deben tomarse con cuidado, ya que el sistema no está presentado como un modelo del funcionamiento real, sino que está haciendo énfasis en el hecho de que varios comportamientos experimentales pueden ser debidos a un fenómeno de sincronismo. Por último, si el sincronismo fuera el máximo responsable de estos comportamientos, creemos que esta representación puede ser útil para entender e interpretar estas observaciones, presentando un interesante punto de partida para estudiarlos.
